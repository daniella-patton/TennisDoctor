# Function to create and Save Basic Data

```python
# Basic Data - name, sex, age, for the top 200 players from 2010 - 2020
def basic_data(soup, year):
    name = soup.title
    names.append(name.text[:-18])
    years.append(str(year))
    for i, table in enumerate(soup.select('table')):
            #print(i, table['class'])
            # Basic Data
            if table['class'] == ['plDetail']:
                rows = table.findAll('div', {'class':'date'})
                for row in rows:
                    row = row.text
                    print(row)
                    # Country
                    if "Country" in row: 
                        countries.append(row[9:])
                    # Birthday
                    if "Age" in row:
                        birthday = row
                        birthday = (birthday[birthday.find("(")+1:birthday.find(")")])
                        birthdates.append(birthday)
                    #Singles Rankings
                    if "Current/Highest rank - singles" in row:
                        sin_ranks = row
                        cur_rank = sin_ranks.split(' / ')[0]
                        cur_rank = cur_rank.split('singles: ')[1]
                        sin_cur_ranks.append(cur_rank)
                        hi_rank = sin_ranks.split(' / ')[1]
                        sin_high_ranks.append(hi_rank)
                    # Doubles Ranking
                    if "Current/Highest rank - doubles" in row:
                        dub_ranks = row
                        cur_rank = dub_ranks.split(' / ')[0]
                        cur_rank = cur_rank.split('doubles: ')[1]
                        dub_cur_ranks.append(cur_rank)
                        hi_rank = dub_ranks.split(' / ')[1]
                        dub_high_ranks.append(hi_rank)
                    # Sex
                    if "Sex" in row:
                        s = row.split('Sex: ')
                        sex.append(s[1])
                    # Handed
                    if "Plays" in row:
                        h = row
                        h = h.split('Plays: ')[1]
                        hand.append(h)
                    # Height & Weight
                    if "Height / Weight" in row:
                        if "Weight" in row:
                            hw = row[17:]
                            h = hw.split(' / ')[0]
                            w = hw.split(' / ')[1]
                        else: 
                            h = row[7:]
                            w = 'N/A'
                        height.append(h)
                        weight.append(w)
                #if table['class'] == ['plDetail']:
    if len(names) != len(birthdates):
        birthdates.append('N/A')
    if len(names) != len(countries):
        countries.append('N/A')
    if len(names) != len(sin_cur_ranks):
        sin_cur_ranks.append('N/A')
    if len(names) != len(sin_high_ranks):
        sin_high_ranks.append('N/A')
    if len(names) != len(dub_cur_ranks):
        dub_cur_ranks.append('N/A')
    if len(names) != len(dub_high_ranks):
        dub_high_ranks.append('N/A')
    if len(names) != len(sex):
        sex.append('N/A')    
    if len(names) != len(height):
        height.append('N/A')
    if len(names) != len(weight):
        weight.append('N/A')
    if len(names) != len(years):
        years.append('N/A')
```

# Injury Data Function

```python
# Extracting and saving player injury in a df
def injury_data(soup):   
    start_injury_date = []
    end_injury_date = []
    tournament = []
    reason = []

    for i, table in enumerate(soup.select('table')):
            # Injury Data        
            if table['class'] == ['result', 'flags', 'injured']:
                rows = table.find_all('tr')
                for row in rows:
                        cols=row.find_all('td')
                        cols=[x.text.strip() for x in cols]
                        if len(cols) == 3:
                            dates = cols[0]
                            start = dates.split(' - ')[0]
                            end = dates.split(' - ')[1]
                            start_injury_date.append(start)
                            end_injury_date.append(end)
                            tournament.append(cols[1])
                            reason.append(cols[2])
                        else:
                            print('length is not 3')

    # Saving the Results as a csv file
    df = pd.DataFrame({'Start_Date':start_injury_date,'End_Date':end_injury_date,
                       'Tournament':tournament,'Reason':reason})
    name = soup.title
    csv = name.text[:-18] + '_Injury_Record.csv'
    df.to_csv(csv, index=False, encoding='utf-8')
```

# Player Tournament Results Function

```python
# Extracting the Player Tournament Results
def tourn_res(soup): 
    tournament = []
    matchtype = []
    opponent = []
    years = []
    date = []
    rounds = []
    results = []
    h = []
    a = []
    # iterate through different years of play
    for j in range(0,13):
        year = 2020 - j
    # need to make sure that play took play that year
        # if exists, collect data, else == skip
        id_type = 'matches-' + str(year) + '-1-data'
        test_exists = soup.find('div', {'id':id_type})
        if test_exists is None:
            print(str(year)+ ' Does not Exist!')
        else:
            print(str(year)+ ' Exists')
        # to iterate through all sing/dubs/mixed dub play in year
            for i in range(1,4): 
                id_type = 'matches-' + str(year) + '-' + str(i) + '-data'
                table = soup.find('div', {'id':id_type})
                if i == 1:
                    mt = 'singles'
                if i == 2:
                    mt = 'doubles'
                if i== 3:
                    mt = 'mixed doubles'

                rows = table.find_all('tr')
                for row in rows:
                    cols=row.find_all('td')
                    cols=[x.text.strip() for x in cols]
                    if len(cols) == 1:
                        tourn = cols[0]
                    else:
                        tournament.append(tourn)
                        date.append(cols[0])
                        opponent.append(cols[2])
                        rounds.append(cols[3])
                        results.append(cols[4])
                        h.append(cols[5])
                        a.append(cols[6])
                        matchtype.append(mt)
                        years.append(str(year))

        # Saving the Results as a csv file
    df = pd.DataFrame({'Year':years,'Tournament':tournament,'Match_type':matchtype,'Date':date,'Opponent':opponent,
                      'Round':rounds,'Result':results,
                      'H':h,'A':a})
    name = soup.title
    csv = name.text[:-18] + '_Record.csv'
    df.to_csv(csv, index=False, encoding='utf-8')
```
